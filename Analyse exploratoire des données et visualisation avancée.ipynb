```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

df = pd.read_csv(r"C:\Users\ayoub\Downloads\Mall_Customers.csv")
print(df.head())
print(df.info())
df.describe()
```

```python
df.isnull().sum()
sns.heatmap(df.isnull(),cbar=False)
plt.title("valeurs manquantes")
plt.show()

```

```python
sns.boxplot(x=df['Age'])
plt.title("Boxplot de l'age")
plt.show()
```

```python
print(df['Gender'].unique()) 
df['Gender'] = df['Gender'].map({'Male': 0, 'Female': 1})
df["Age"] = df["Age"].fillna(df["Age"].mean())
df.head()
```

```python
sns.countplot(x='Gender', data=df)
plt.title('Répartition des clients selon le genre')
plt.xlabel('Genre')
plt.ylabel('Nombre de clients')
plt.show()
```

```python
sns.histplot(df['Age'], kde=True, bins=20)
plt.title("Distribution de l'âge des clients")
plt.xlabel("Age")
plt.ylabel("Fréquence")
plt.show()
```

```python
sns.boxplot(y='Annual Income (k$)', data=df)
plt.title('Distribution du revenu annuel')
plt.ylabel('Revenu annuel (k$)')
plt.show()
```

```python
sns.boxplot(x='Gender', y='Annual Income (k$)', data=df)
plt.title('Revenu annuel selon le genre')
plt.show()
```

```python
sns.violinplot(x='Gender', y='Spending Score (1-100)', data=df)
plt.title('Score de dépenses par genre')
plt.show()
```

```python
df['age_group'] = pd.cut(df['Age'], bins=[18, 25, 35, 45, 60, 100], labels=['18-25', '26-35', '36-45', '46-60', '60+'])

sns.boxplot(x='age_group', y='Spending Score (1-100)', data=df)
plt.title("Score de dépenses par tranche d'âge")
plt.show()
```

```python
sns.scatterplot(x='Annual Income (k$)', y='Spending Score (1-100)', hue='Gender', size='Age', data=df)
plt.title("Score de dépenses vs Revenu annuel")
plt.show()
```
